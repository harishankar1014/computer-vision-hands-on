```python
import numpy as np
import cv2 as cv
from PIL import Image
```

```python
img = cv.imread('dataset/taj.jpg')
gray= cv.cvtColor(img,cv.COLOR_BGR2GRAY)
```

```python
sift = cv.SIFT_create()
kp = sift.detect(gray,None)
img_sift = img.copy()
```

```python
img_sift=cv.drawKeypoints(gray,kp,img_sift)
cv.imwrite('sift_keypoints.jpg',img_sift)
img_sift=cv.drawKeypoints(gray,kp,img_sift,flags=cv.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
```

```python
Image.fromarray(img_sift)
```

```python
#Scaling 
scaled_image = cv.resize(img, (1050, 1610))
Image.fromarray(scaled_img)
```

```python
#rotated image
rotated_image = cv.rotate(img, cv.ROTATE_90_CLOCKWISE)
Image.fromarray(rotated_image)
```

```python
#affine transformation
rows, cols, ch = img.shape
pts1 = np.float32([[50, 50],
                   [200, 50],
                   [50, 200]])
 
pts2 = np.float32([[10, 100],
                   [200, 50],
                   [100, 250]])
M = cv.getAffineTransform(pts1, pts2)
affine_image = cv.warpAffine(img, M, (cols, rows))
Image.fromarray(affine_image) 
```

```python
#Perspective transformation
rows, cols, ch = img.shape
pts1 = np.float32([[0,0],
                   [50, 50],
                   [200, 50],
                   [50, 200]])
 
pts2 = np.float32([[0,0],
                   [10, 100],
                   [200, 50],
                   [100, 250]])
P = cv.getPerspectiveTransform(pts1, pts2)
perspective_image = cv.warpPerspective(img, P, (cols, rows))
Image.fromarray(perspective_image)
```

```python
scaled_gray= cv.cvtColor(scaled_image,cv.COLOR_BGR2GRAY)
rotated_gray = cv.cvtColor(rotated_image,cv.COLOR_BGR2GRAY)
affine_gray= cv.cvtColor(affine_image,cv.COLOR_BGR2GRAY)
perspective_gray= cv.cvtColor(perspective_image,cv.COLOR_BGR2GRAY)
```

```python
#Scaled image sift
sift = cv.SIFT_create()
kp = sift.detect(scaled_gray,None)
scaled_img_sift=cv.drawKeypoints(scaled_gray,kp,scaled_image)
cv.imwrite('sift_keypoints.jpg',scaled_img_sift)
scaled_img_sift=cv.drawKeypoints(scaled_gray,kp,scaled_img_sift,flags=cv.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
Image.fromarray(scaled_img_sift)
```

```python
#Rotated image sift
sift = cv.SIFT_create()
kp = sift.detect(rotated_gray,None)
rotated_img_sift=cv.drawKeypoints(rotated_gray,kp,scaled_image)
cv.imwrite('sift_keypoints.jpg',rotated_img_sift)
rotated_img_sift=cv.drawKeypoints(rotated_gray,kp,rotated_img_sift,flags=cv.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
Image.fromarray(rotated_img_sift)
```

```python
#Affine image sift
sift = cv.SIFT_create()
kp = sift.detect(affine_gray,None)
affine_img_sift=cv.drawKeypoints(affine_gray,kp,scaled_image)
cv.imwrite('sift_keypoints.jpg',rotated_img_sift)
affine_img_sift=cv.drawKeypoints(affine_gray,kp,affine_img_sift,flags=cv.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
Image.fromarray(affine_img_sift)
```

```python
#Perspective image sift
sift = cv.SIFT_create()
kp = sift.detect(perspective_gray,None)
perspective_img_sift=cv.drawKeypoints(perspective_gray,kp,scaled_image)
cv.imwrite('sift_keypoints.jpg',rotated_img_sift)
perspective_img_sift=cv.drawKeypoints(perspective_gray,kp,perspective_img_sift,flags=cv.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
Image.fromarray(perspective_img_sift)
```
