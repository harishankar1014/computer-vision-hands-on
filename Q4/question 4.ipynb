# 1. Edge based segmentation

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage.feature import canny
from skimage import data,morphology
from skimage.color import rgb2gray
import scipy.ndimage as nd
plt.rcParams["figure.figsize"] = (12,8)
%matplotlib inline
```

```python
# load images and convert to grayscale
```

```python
import skimage
```

```python
im1= skimage.io.imread('tvm1.jpg')
im2= skimage.io.imread('tvm2.jpg')
im3= skimage.io.imread('tvm3.jpg')
im4= skimage.io.imread('tvm4.jpg')
```

```python
#converting to grey scale
```

```python
im1g = rgb2gray(im1)
im2g = rgb2gray(im2)
im3g = rgb2gray(im3)
im4g = rgb2gray(im4)
```

```python

edge1 = canny(im1g)  # applying edge segmentation
edge2 = canny(im2g)
edge3 = canny(im3g)
edge4 = canny(im4g)
plt.imshow(edge1, interpolation='gaussian') # plotting figure
plt.title('Image1')
```

```python
plt.imshow(edge2, interpolation='gaussian') # plotting figure
plt.title('Image2')
```

```python
plt.imshow(edge3, interpolation='gaussian') # plotting figure
plt.title('Image3')
```

```python
plt.imshow(edge4, interpolation='gaussian') # plotting figure
plt.title('Image4')
```

# 2. Region Based segmentation

```python
region1= nd.sobel(im1g)
plt.imshow(region1)
```

```python
region2 = nd.sobel(im2g)
plt.imshow(region2)
```

```python
region3 = nd.sobel(im3g)
plt.imshow(region3)
```

```python
region4= nd.sobel(im4g)
plt.imshow(region4)
```

```python
# the problem here is contrast is not up to a fairly distinguishable level, we are gonna add further distinguishability
```

```python
from skimage.segmentation import watershed
from skimage import segmentation
```

```python
markers = np.zeros_like(im1g)
markers[im1g < 0.1171875] = 1 # 30/255
markers[im1g > 0.5859375] = 2 # 150/255
 
plt.imshow(markers)
plt.title('markers')
 

segment1 = segmentation.watershed(region1, markers)
 
plt.imshow(segment1)
plt.title('pic1')
```
